```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
from datetime import datetime
import pandas as pd
import requests
import re
import os
```

```python
keyword = 'air+quality'
crawl_base_url = f"https://paperswithcode.com/search?q_meta=&q_type=&q={keyword}"
```

```python
metadata = []
# Crawl the first 2 pages
for page in range(1,2):
  # Add page number and headers info to crawl, the headers object
  # identifies the user agent (i.e., the web browser or application) making the request.
  req = Request(url=f"{crawl_base_url}&page={page}", headers={'User-Agent': 'Mozilla/5.0'})

  # Get the response and convert to raw HTML
  html = urlopen(req).read().decode('utf-8')

  # Create soup object out of raw HTML, this will create parsed HTMl
  # tree to extract the data
  soup = BeautifulSoup(html, features="lxml")

  # Find all function receives 2 parameters, the first one is HTMl tag
  # we want to get, the second is some extra information
  articles = soup.find_all('div', {'class': 'row infinite-item item paper-card'})

  # loop through the divs and extract data from elements inside
  for article in articles:
    # Info variable will be used to save data of a paper
    info = {}
    # Get general information
    # We call find method to find child elements inside the article element
    # It receives 2 parameters: first is the tag name and second optional
    # parameter is the class name of element we want to get
    info["title"] = article.find('h1').get_text()
    info["stars"] = article.find('div', {'class': 'entity-stars'}).get_text()
    info["image_url"] = article.find('div', {'class': 'item-image'})['style']

    ############################# YOUR CODE GOES HERE #########################


    ############################# YOUR CODE GOES HERE #########################


    # Add data to the metadata arrays contain all crawled data
    metadata.append(info)
```
